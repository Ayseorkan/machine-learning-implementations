```python
y=df_train.SalePrice
X=df_train.drop("SalePrice", axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y,train_size=0.8, random_state=0)
```

```python
categorical_cols=[cname for cname in X_train.columns if X_train[cname].nunique()<10 and X_train[cname].dtype=="object"]
```

```python
len(categorical_cols)
```

```python
numerical_cols=[cname for cname in X_train.columns if X_train[cname].dtype in ["int64", "float6]"]]
```

```python
len(numerical_cols)
```

```python
my_cols=categorical_cols+numerical_cols
X_train=X_train[my_cols].copy()
X_valid=X_valid[my_cols].copy()
X_test=df_test[my_cols].copy()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
numerical_transformer=Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

```python
categorical_transformer=Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
from sklearn.compose import ColumnTransformer
preprocessor=ColumnTransformer(transformers=[
    ("num", numerical_transformer, numerical_cols),
    ("cat", categorical_transformer, categorical_cols)
])
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model=RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
my_pipeline=Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", rf_model)
])
```

```python
my_pipeline.fit(X_train, y_train)
```

```python
valid_preds=my_pipeline.predict(X_valid)
```

```python
from sklearn.metrics import mean_absolute_error
print("MAE: ", mean_absolute_error(y_valid, valid_preds))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores= -1*cross_val_score(my_pipeline, X, y, cv=5, scoring="neg_mean_absolute_error")
scores.mean()
```

```python
test_preds = my_pipeline.predict(X_test)
```

```python
submission = pd.DataFrame({
    "Id": df_test.index,
    "SalePrice": test_preds
})

submission.to_csv("submission.csv", index=False)
```
